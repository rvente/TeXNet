```python
vd = VisualizeDir('I2L-STRIPS')
unmatched, missing = pub.get_unmatched_images('I2L-STRIPS/eval_images/rendered_images')
last_step = vd.get_steps()[1][-1]
vs = VisualizeStep(vd, 'test', last_step)
df_strs = vs.strs( 'y', 'predicted_ids', mingle=False, trim=True, wrap_strs=True, keys=['image_name'])
df_strs['image_name_trunc'] = df_strs.image_name.str.replace('_basic.png', '.png')
df_strs = df_strs[~df_strs.image_name_trunc.isin(unmatched+missing)].sort_values(by='predicted_ids_len', ascending=False)
print(df_strs.shape)
display(df_strs.iloc[:10])
```
